```python
import numpy as np
import matplotlib.pyplot as plt
```

We take $\mathcal{V}^2/\hbar^2$ = 1.
We define a time array (scaled by $\Omega_R$):

```python
tau = np.linspace(0, 25, 1000)
```

We are going to plot various detuning values, we take $\Delta$ = 0, 0.5 and 1.0.

```python
deltas = [0, 0.5, 1.0]
styles = ['-', ':', '-.']
labels = [r'$\Delta = 0$', r'$\Delta = 0.5$', r'$\Delta = 1$']
```

We procced with the plotting.

```python
plt.figure(figsize=(8, 5))

for delta, style, label in zip(deltas, styles, labels):
    Omega_R = np.sqrt(1 + delta**2)
    P_e = 1 / (1 + delta**2) * np.sin(0.5 * Omega_R * tau)**2
    plt.plot(tau, P_e, style, label=label)

plt.xlabel(r'$\Omega_R t$', fontsize=14)
plt.ylabel(r'$P_e(t)$', fontsize=14)


plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)


plt.legend(fontsize=13) 

plt.ylim(0, 1.05)
plt.xlim(0, 25)
plt.tight_layout()
plt.show()
```
